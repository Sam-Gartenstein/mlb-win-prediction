### Exporting Data

```python
# Collect existing season DataFrames into a dict automatically
seasons = {
    year: df for year in range(2021, 2026)
    if (df := globals().get(f"season_{year}")) is not None
}

# Create subfolder: data/raw_season_data
outdir = Path("data") / "raw_season_data"
outdir.mkdir(parents=True, exist_ok=True)

for year, df in seasons.items():
    df.to_csv(outdir / f"season_{year}.csv", index=False, encoding="utf-8")
    print(f"[saved] {outdir / f'season_{year}.csv'}")
```
